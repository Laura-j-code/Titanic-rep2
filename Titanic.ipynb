# Machine Learning

```python
tt_copy= titanic.copy()
dummed_data= pd.get_dummies(data=tt_copy,columns=['Embarked'])
dummed_data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
enc= LabelEncoder()
dummed_data['enc_sex']= enc.fit_transform(dummed_data['Sex'])
dummed_data.drop(columns='Sex',inplace=True)
dummed_data

```

```python
from sklearn.preprocessing import OneHotEncoder
enc2= OneHotEncoder()

hot_enc= enc2.fit_transform(tt_copy['Embarked'].values.reshape(-1,1)).toarray()
hot2= pd.DataFrame(hot_enc,columns=['None','C','Q','S'])
hot3= pd.concat([dummed_data,hot2],axis=1)
hot3.drop(columns=['Embarked_C','Embarked_Q','Embarked_S'],inplace=True)
hot3
```

```python
num_df = dummed_data.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(num_df.corr(), annot=True,)
plt.show()
```

# Machine learning

### Logistic regression model performance

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X= hot3.drop('Survived',axis=1)
y= hot3['Survived']

X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=3)

print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))


X_train_scaler=scaler.fit_transform(X_train)
X_test_scaler=scaler.fit(X_test)


from sklearn.linear_model import LogisticRegression
lrg= LogisticRegression( C=1.0,penalty='l2',solver='lbfgs',max_iter=200 )
'''Fitting the model'''

model= lrg.fit(X_train,y_train)
model
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report 

y_pred= model.predict(X_test)

print('accuracy_score :',accuracy_score(y_pred,y_test))
print('confusion_matrix:',confusion_matrix(y_pred,y_test))
print('classification_report:',classification_report(y_pred,y_test))


from sklearn.ensemble import RandomForestClassifier

```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [0.01, 0.1, 1],
        'max_iter': [10, 100, 1000, 2000],
        'class_weight': ['balanced', None]
    },
    {
        'solver': ['lbfgs', 'newton-cg', 'sag'],
        'penalty': ['l2', 'none'],
        'C': [0.01, 0.1, 1],
        'max_iter': [10, 100, 1000, 2000],
        'class_weight': ['balanced', None]
    },
    {
        'solver': ['saga'],
        'penalty': ['l1', 'l2', 'none'],
        'C': [0.01, 0.1, 1],
        'max_iter': [10, 100, 1000, 2000],
        'class_weight': ['balanced', None]
    },
    {
        'solver': ['saga'],
        'penalty': ['elasticnet'],
        'l1_ratio': [0.0, 0.25, 0.5, 0.75, 1.0], 
        'C': [0.01, 0.1, 1],
        'max_iter': [10, 100, 1000, 2000],
        'class_weight': ['balanced', None]
    }
]


grid_search = GridSearchCV(
    LogisticRegression(random_state=11),
    param_grid=param_grid,
    scoring="accuracy",
    cv= 5,
    n_jobs=-1,
    verbose= True
)


best_param = grid_search.fit(X_train,y_train)
y_pred3= best_param.predict(X_test)


print('accuracy_score :', accuracy_score(y_test, y_pred3))
print('confusion_matrix:', confusion_matrix(y_test, y_pred3))
print('classification_report:', classification_report(y_test, y_pred3))
```

### Random forrest model performance

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


rf_model = RandomForestClassifier(random_state=11)

rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

print('accuracy_score :', accuracy_score(y_test, y_pred_rf))
print('confusion_matrix:\n', confusion_matrix(y_test, y_pred_rf))
print('classification_report:\n', classification_report(y_test, y_pred_rf))



from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300], 
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4],      
    'bootstrap': [True, False]               
}

grid_search_rf = GridSearchCV(
    RandomForestClassifier(random_state=11),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search_rf.fit(X_train, y_train)
y_pred_rf = grid_search_rf.best_estimator_.predict(X_test)

# Evaluate
print('Best Parameters:', grid_search_rf.best_params_)
print('Best CV Accuracy:', grid_search_rf.best_score_)
print('accuracy_score :', accuracy_score(y_test, y_pred_rf))
print('confusion_matrix:\n', confusion_matrix(y_test, y_pred_rf))
print('classification_report:\n', classification_report(y_test, y_pred_rf))

```

## XGBoost model performance

```python
import xgboost as xgb

xgb_clf = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=3,
    max_depth=4,
    learning_rate=0.1,
    n_estimators=100,
    random_state=42
)

xgb_clf.fit(X_train, y_train)

# Predictions
y_pred4 = xgb_clf.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred4)
print(f"Accuracy: {accuracy:.4f}")

```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': np.linspace(0.7, 1.0, 4),
    'colsample_bytree': np.linspace(0.7, 1.0, 4),
    'gamma': [0, 0.1, 0.2, 0.3]
}

# Randomized Search
random_search = RandomizedSearchCV(
    estimator=xgb_clf,
    param_distributions=param_dist,
    n_iter=20,  # Number of random combinations to try
    scoring='accuracy',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

# Best parameters and model evaluation
print("Best Parameters:", random_search.best_params_)
best_model = random_search.best_estimator_

y_pred5 = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred5)
print(f"Test Accuracy: {accuracy:.4f}")
```
